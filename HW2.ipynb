```python
# Question 1 - i

"""
Generates A password of 8 charactes
Date: 26/09/2022
Took 20 runs to finish program
"""

#Importing libraries and creating variables
import random
import string
password = [] #this is to later shuffle all strings

#------
#Generating 2 random integers
int1 = str(random.randint(0,9)) #generates first integer
int2 = str(random.randint(0,9)) #generate second integer
password.extend((int1,int2)) #adding it to list
print ("Integers:",int1+int2) #this is a test run to check if it works
print ("Password update:",password)

"""Reason I didn't do it from 00-99 as any numbers that start with 0, the 0 would be removed.
   I also changed it to string so that it becomes easier to add it to the rest later. 
   I also just dicovered there was a specific function for numbers in the string library."""

#-----
#Generating 2 random Upper Alphanumeric Characters
upper1 = random.choice(string.ascii_letters).upper() #generate first upper letter
upper2 = random.choice(string.ascii_letters).upper() #generate second upper letter
password.extend((upper1,upper2)) #adding it to list
print ("Upper Letters:", upper1+upper2) #this is a test run to check if it works
print ("Password update:",password)

#-----
#Generating 2 random Lower Alphanumeric Characters
lower1 = random.choice(string.ascii_letters).lower() #generate first lower letter
lower2 = random.choice(string.ascii_letters).lower() #generate second lower letter
password.extend((lower1,lower2)) #adding it to list
print ("Lower Letters:",lower1+lower2) #this is a test run to check if it works
print ("Password update:",password)

#-----
#Generating 2 random symbols
symbol1 = random.choice(string.punctuation) #generate symbol
symbol2 = random.choice(string.punctuation) #generate symbol
password.extend((symbol1,symbol2)) #adding it to list
print ("Symbols:",symbol1+symbol2) #this is a test run to check if it works
print ("Password update:",password)

#-----
#Suffling and combining
print ("Password before shuffle:",''.join(password)) #this is a test run to check if it works
shuffled_password = random.shuffle(password) #shuffles password and changes the original list
print ("Password after shuffle:",''.join(password)) #this is a test run to check if it works

#-----
#Final Output
print ("\nRecommended Password:",''.join(password)) #Final output
```

```python
#Question 3 - ii

"""
Date: 29/09/2022
Took 13 runs to finish program
"""

"""
PLAN:
Tupples can be changed in to a list, edited, then changed back into a tuple
"""

#Preparing variables used in the code
tups_list = [(10, 20, 40), (40, 50, 60), (70, 80, 90)]
number_replace = int(input())
new_tups_list = []


#Main Body
print ("The List with Tuples:",tups_list) #to check what was inputed
print ("Number to be placed instead:",number_replace) #to check what number we're replacing it by

for replace in tups_list:
    print ("\nOriginal Tupple:",replace) #Printing to check if it works

    tups_to_list = list(replace) #Changes the tupple into a list
    print ("Original List Tupple:",tups_to_list) #Printing to check if it works

    tups_to_list[-1] = number_replace #replaces the last number of the tupple with chose number
    print ("Replaced List Tupple:",tups_to_list) #Printing to check if it works

    back_to_tups = tuple(tups_to_list) #Reverts the list back to tupple
    print ("Replaced Tupple:",back_to_tups) #Printing to check if it works

    new_tups_list.append(back_to_tups) #Add it to a seperate list
    print ("The list so far:", new_tups_list) #Printing to check if it works

#Final Output
print ("\n",new_tups_list) #Final Output
```

```python
#Question 3 - iv

"""
Date: 29/09/2022
Took 25 runs to finish program
"""

"""
PLAN:

Number <-- Number inputed
For i in range (1,Number):
    If number % i with no remainders
        True: Add it to list
        False: Continue

Sums List
If sum == Number
    True: Perfect Number
    False: Not Perfect Number
"""

#Variables to be used
num = int(input()) #Number inputed
list_of_factors = []

#Main Body
for factor in range (1,num): #starts from one until the number, but excluding it
    print ("\nChecking if '{}' is a factor".format(factor)) #to keep track of the process
    if num%factor == 0: #Checks if there is a remainder when dividing
        list_of_factors.append(factor) #If there isn't adds the factor to the list
        print ("The list so far:",list_of_factors) #printing to check if it is working correctly
    else: #if it isn't a factor
        print ("'{}' is not a factor".format(factor)) #printing to check if it works correctly

sum_of_factors = sum(list_of_factors) #sums it up

#Final output and checking if it is a perfect number
if sum_of_factors == num:
    print ("\n'{}' is a perfect number".format(num)) #if it is a perfect number
else:
    print ("\n'{}' is not a perfect number".format(num)) #if it isn't a perfect number
```

```python
#Question 4 - ii

"""
Date: 29/09/2022
Took [] runs to finish program
"""

"""
PLAN:

Makes all small caps

Revrses the letters by using a for loop in the opposite direction
Compares the original to the reversed

Are they the same:
    True: Palindrome
    False: Not Palindrome
"""

#Creating Variables
original_word = input()
reversed_word_list = []

#Main Body
print ("Original Word:", original_word) #prints it to check it

lower_original_word = original_word.lower() #makes it all small case
print ("\nSmall Original Word:",lower_original_word) #prints it to check it

range_of_word = (len(original_word)*-1)-1
print ("\nRange: {}\n".format(range_of_word))

for letter in range (-1,range_of_word,-1): #Loop to reverse the word
    print ("Position: {}\nLetter: {}".format(letter,lower_original_word[letter])) #To check if it is reversed
    reversed_word_list.append(lower_original_word[letter]) #adds it to a list
    print ("List Update:", reversed_word_list) #To check it is working

reversed_word = ''.join(reversed_word_list)
print ("Reversed word:",reversed_word)

if original_word==reversed_word:
    print ("'{}' is a Palindrome".format(original_word))
else:
    print ("'{}' is not a Palindrome".format(original_word))
```
